```python
"""
Q.1.  What is a database? Differentiate between SQL and NoSQL 
databases.

Solution:

A database is a structured collection of data that is
organized and stored for efficient retrieval and manipulation.
Databases are essential in many applications and systems,
serving as a means to store, manage, and access data.
They come in various forms and are used in a wide range of 
contexts, from simple personal record-keeping to large-scale
enterprise systems.

Here's a differentiation between SQL (Relational) and NoSQL 
databases:

1. SQL (Relational) Databases:
Data Structure: SQL databases are also known as relational 
databases because they use a tabular structure to store data. 
Data is organized into tables with rows and columns, where
each row represents a record, and each column represents an
attribute or field.

2. Schema: SQL databases have a fixed schema, which means
that you define the structure of the database, including the
tables, columns, and data types, in advance. Changes to the
schema can be complex and require careful planning.

3. Query Language: SQL (Structured Query Language) is used to 
interact with relational databases. SQL is a powerful and 
standardized language for querying, updating, and manipulating 
data.

4. ACID Transactions: SQL databases typically support ACID 
(Atomicity, Consistency, Isolation, Durability) transactions, 
ensuring data integrity and consistency. This is essential for
applications that require strong data consistency and 
reliability, such as financial systems.

5. Scalability: SQL databases are traditionally considered to
be vertically scalable, meaning that you can increase their 
performance by upgrading the hardware resources (CPU, memory,
etc.) of a single server. This can be costly and has practical
limits.

6. Use Cases: SQL databases are well-suited for applications
where data integrity and complex queries are paramount, such
as financial systems, e-commerce platforms, and applications
that require complex joins and transactions.

NoSQL Databases:

1. Data Structure: NoSQL databases use various data models,
including document, key-value, column-family, and graph. Each 
data model has a different way of organizing and storing data,
but they all allow for flexible, schema-less data storage.

2. Schema: NoSQL databases are schema-less or schema-flexible,
meaning you can add fields to records/documents without the
need to modify a predefined schema. This flexibility is
beneficial for applications with evolving data requirements.

3. Query Language: NoSQL databases have their own query
languages tailored to their data models. The query languages
are typically less expressive than SQL, but they are optimized
for specific use cases.

4. ACID Transactions: NoSQL databases vary in their support 
for ACID transactions. Some provide strong consistency,
while others prioritize scalability and offer eventual 
consistency.

5. Scalability: NoSQL databases are typically designed to be 
horizontally scalable, meaning you can add more servers to
distribute the load. This makes them well-suited for
applications with high read and write loads, such as 
social media platforms and real-time analytics.

6. Use Cases: NoSQL databases are commonly used in applications
that require high scalability, low-latency access, and can 
tolerate some level of data inconsistency. Examples include
content management systems, real-time analytics, and IoT 
applications.
"""
```
