```python
"""
Q.5. Explain Primary Key and Foreign Key.

Solution:

Primary Key and Foreign Key are fundamental concepts in
relational databases, and they are used to establish 
relationships between tables and ensure data integrity.

Primary Key: A Primary Key is a column (or a set of columns)
in a relational database table that uniquely identifies each
row or record in the table. It serves as a unique identifier
for each row, ensuring that no two rows have the same values
in the primary key column(s). 

Here are some key points about primary keys:

Uniqueness: Every value in the primary key column(s) must 
be unique within the table. This uniqueness constraint
ensures that each row can be easily and uniquely identified.

Non-null: The values in the primary key column(s) cannot be
NULL, meaning they must have a value for every row.

Data Integrity: Primary keys help maintain data integrity 
by preventing duplicate or inconsistent data.

Indexes: Database management systems typically create an
index on the primary key column(s) to enhance query
performance when searching for specific rows.

Auto-incrementing: Primary keys are often auto-incrementing,
meaning that the database system generates a unique value
for the primary key for each new row automatically.

Example of a Primary Key:
In a "Students" table, the "student_id" column can be the
primary key. This ensures that each student has a unique
identifier, and no two students share the same "student_id."

Foreign Key:
A Foreign Key is a column (or a set of columns) in one table
that is used to establish a link between the data in two tables.
The foreign key in one table is linked to the primary key in
another table, creating a referential relationship. Here are
some key points about foreign keys:

Referential Integrity: Foreign keys help maintain referential 
integrity by ensuring that the data in one table corresponds
to data in another table. It enforces data consistency and 
relationships between tables.

Relationships: Foreign keys establish relationships between 
tables, allowing you to connect data in meaningful ways.
For example, you can link a "student_id" in a "Grades" 
table to the "student_id" in a "Students" table, creating 
a link between students and their grades.

Cascading Actions: Foreign keys can define actions to be 
taken when data in the referenced table (the one with the
primary key) is modified. These actions can include cascading 
deletes or updates, maintaining data integrity.

Example of a Foreign Key:
In a database with "Students" and "Grades" tables, the 
"student_id" column in the "Grades" table can be a foreign
key linked to the "student_id" column in the "Students" table.
This establishes a relationship between a student's grades 
and the student's information.
"""
```
